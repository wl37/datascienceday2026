# Targeted Promotion Campaign

## Scenario: BrightBite Cafe

BrightBite is a neighborhood cafe with limited $2000 marketing budget.

- Off-peak hour 10-11AM has few customers;
- Short on staff in early morning, customers complained about long waiting time

**Questions to consider**
- What's the purpose of the marketing campaign? 
- How should you design the promotion offer to match your campaign goal? (e.g. offer $ discount? BOGO 50% off?)
- How much does each promotion offer cost? What's the max promotion offer you can run with the budget?
- How can BrightBite send the **right promotion to the right customer**?

## What Must a Good Targeted Promotion Include?

A useful promotion decision usually depends on four elements:

1. **Who** is the customer?
2. **What** offer should they receive?
3. **When** should we send it?
4. **Which channel** should we use (app, email, SMS)?

## What Data Can Help to Make Decisions?

Suppose BrightBite can collect:

- Purchase history (items, amount, date/time)

- Promotion history (what offer was sent, and response)

### Discussion

## How Could ML Help? 

Business question:
- "Who should receive a 20% pastry coupon this Friday?"

Possible ML formulations:
- **Binary classification:** Will customer redeem coupon? (Yes/No)
- **Ranking:** Which customers are most likely to respond?
- **Recommendation:** Which offer fits each customer best?

Pick one formulation and explain why it matches business goals better than the others.

## Key Features: What Signals Help Predict Response?

Potential features (inputs):

- Visit frequency in last 30 days
- Average spending per visit
- Favorite category (coffee, bakery, meals)
- Time-of-day preference (morning/afternoon/evening)
- Days since last purchase
- Past coupon redemption rate

### Discuss

- Which features are likely strong predictors?
- How can you get this feature (feature engineering)
- Which features could be unfair or risky to use?
- How might feature choices influence customer experience?

## Target Variable: What Exactly Are We Predicting? 

Examples of target variables:

- **RedeemWithin7Days**: Did the customer use the coupon in 7 days? (0/1)
- **SpendAfterPromotion**: How much did customer spend after receiving offer?
- **NextBestOffer**: Which offer type leads to highest engagement?

Choose one target variable and define how you would measure it clearly from real data.

## (Optional) Solution Approach: How Would We Build the ML System?

### A practical ML workflow

1. Define objective and evaluation metric (e.g., redemption rate uplift)
2. Prepare training data (clean, label, split train/validation/test)
3. Start with simple models (logistic regression, decision tree)
4. Compare with stronger models (random forest, gradient boosting)
5. Validate with business constraints (budget, fairness, explainability)


### Responsible AI: What Could Go Wrong?

Targeted promotion can create value, but also risks:

- Privacy concerns (collecting too much personal data)
- Over-targeting (customers feel manipulated)
