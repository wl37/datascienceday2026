# Targeted Promotion Campaign

### Meet BrightBite Cafe

Imagine you're **Maya**, the owner of a neighborhood cafe called **BrightBite**. You opened a year ago, and things are going pretty well — regulars love your pastries, the coffee is solid, and mornings are busy.

But lately you've been staring at the numbers, and two things worry you.

**Problem 1 — The afternoon ghost town.**

Every day from 2 to 5 PM, the cafe is nearly empty. Three or four customers an hour. Maya is still paying rent, electricity, and a barista — but the register barely moves. She can't just close early; a cafe that shuts down at 2 PM loses credibility in the neighborhood. So she watches money drain every afternoon.

**Problem 2 — Customers who vanish.**

Six months ago Maya launched a simple loyalty app. Looking at the data, she discovers something painful: **4 out of 10 first-time visitors never come back.** She worked hard to get them through the door — and they disappeared.

Maya has saved up **$2,000** for her first real marketing campaign. It's enough for one well-designed push — but not enough to waste.

> *If you were Maya, how would you spend that $2,000?*

## Your Turn — Design a Promotion

Before we think about data or models, let's start with business instinct. With your group, answer these questions:

- **What is the goal?** Fill the afternoon hours? Win back lapsed customers? Both?
- **What kind of offer would work?** A dollar-off discount? Buy-one-get-one 50% off? A free pastry with any afternoon drink?
- **How much does each offer actually cost?** If a pastry costs $1.50 to make and you give it away free, that's 1,333 pastries on a $2,000 budget. Is that a lot — or a little?

**Design one promotion** you think would work best for BrightBite — and be ready to explain *why*.

## The Right Offer to the Right Person

Here's the catch: blasting the same coupon to everyone wastes money. Think about it —

- A college student who already drops by at 3 PM for a study session doesn't need an afternoon discount. You'd be giving away margin for nothing.
- A morning regular who grabs coffee at 7:30 AM and never comes back later? They *might* respond to an afternoon pastry deal — or they might just ignore it.
- A customer who visited once three weeks ago and never returned? They probably need a bigger nudge than a 10% coupon.

A good targeted promotion answers four questions:

| Question | Example |
|:---------|:--------|
| **Who** is the customer? | Morning regular, afternoon studier, lapsed visitor |
| **What** offer should they get? | 20% off pastry, free drink upgrade, BOGO |
| **When** should we send it? | Thursday evening for a Friday afternoon visit? |
| **Which channel** works best? | App notification, email, SMS? |

With your group: **pick one customer type and design a targeting rule** — for example, *"Lapsed customers who haven't visited in 14+ days get a free pastry offer via SMS on Friday morning."*

Which rule do you like best? Why?

## What Data Would You Need?

Suppose BrightBite can collect:

- **Purchase history** — what items each customer bought, how much they spent, and when
- **Basic profile** — age group, student or working adult, neighborhood
- **Promotion history** — what offers were sent before, and whether the customer redeemed them

Before you get excited, think about the hard questions:

- Which of these is **easy** to collect right now? (Hint: what does the loyalty app and point-of-sale system already record?)
- Which data requires **asking the customer for permission**?
- What could go wrong if the data is **incomplete or biased**? (What if only app users are tracked, but half your customers pay cash and never signed up?)

## Could ML Help Here? (Optional)

Let's say Maya asks you a very specific question:

> *"I have 500 customers in my loyalty app and budget for 200 pastry coupons. **Who should get one?**"*

There are several ways to frame this as an ML problem:

- **Binary classification** — predict whether each customer will redeem the coupon (Yes / No)
- **Ranking** — score all 500 customers by likelihood and send coupons to the top 200
- **Recommendation** — figure out which *type* of offer fits each customer best

Each framing leads to a different kind of model, different data needs, and different trade-offs.

**Discuss with your group:** which framing makes the most sense for Maya's situation — and why?

## What Signals Would the Model Use? (Optional)

If *you* were the model, what clues would help you decide? Here are some features BrightBite could compute for each customer:

- **Visit frequency** in the last 30 days
- **Average spending** per visit
- **Favorite category** — coffee, bakery, or meals?
- **Time-of-day preference** — do they come in the morning, afternoon, or evening?
- **Days since last purchase** — are they drifting away?
- **Past coupon redemption rate** — have they used promotions before?

Think about these with your group:

- Which features are likely the **strongest predictors** of whether someone will redeem a coupon?
- How would you actually **compute** each feature from raw purchase records? (This is called *feature engineering* — and it's often the hardest part of a real project.)
- Are any of these features **risky to use**? Could targeting by age group or neighborhood feel unfair to some customers?

## What Exactly Are We Predicting? (Optional)

This sounds obvious — but it's one of the most important decisions in any ML project. The **target variable** defines what "success" means, and getting it wrong means your model optimizes for the wrong thing.

Three options for BrightBite:

| Target Variable | What It Measures | Type |
|:----------------|:-----------------|:-----|
| **RedeemedWithin7Days** | Did the customer use the coupon within a week? | Yes / No |
| **SpendAfterPromotion** | How much did the customer spend after receiving the offer? | Dollar amount |
| **NextBestOffer** | Which offer type leads to the highest engagement? | Category |

Each one sounds reasonable — but they lead to very different models and very different business outcomes.

**Pick one** and think through: how would you measure it clearly from real data? What if a customer redeems the coupon but only buys the discounted item and nothing else — is that a win for Maya?
